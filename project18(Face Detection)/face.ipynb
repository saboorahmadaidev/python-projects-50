```python
import cv2

# Use OpenCV's built-in haarcascade path (portable & clean)
face_cap = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# IP Webcam stream URL
stream_url = "your ip webcam stream url here"  # e.g., "http://

# Connect to mobile camera stream
video_cap = cv2.VideoCapture(stream_url)

if not video_cap.isOpened():
    print("‚ùå Could not connect to IP Webcam stream")
    exit()

print("‚úÖ Connected to IP Webcam")

while True:
    ret, frame = video_cap.read()

    if not ret:
        print("‚ùå Failed to receive frame")
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cap.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    # Draw rectangle around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("üì± IP Webcam Face Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_cap.release()
cv2.destroyAllWindows()
```
